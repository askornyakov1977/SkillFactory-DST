1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# Объявляем функцию которая принимает строку и возвращает урезанную до слова 'образование'
def level_of_education(feature):
    """Функция-преобразования для нового признака 'Образование'

    Args:
        feature (str): значение должно браться из колонки 'Образование и ВУЗ'

    Returns:
        str: урезанное значение от входного до слова 'образование'
    """
    lst = feature.split()
    try:
        i = lst.index('образование')
    except:
        return 'Образование не указано'
    else:
        return ' '.join(lst[:i])
# Создаем с помощью функции-преобразования новый признак 'Образование'
data['Образование'] = data['Образование и ВУЗ'].apply(level_of_education)
# Удаляем колонку 'Образование и ВУЗ'
data = data.drop('Образование и ВУЗ', axis=1)
# Переобразовываем признак 'Образование' в категориальный тип данных
data['Образование'] = data['Образование'].astype('category')
# Проверим полученную категорию, с помощью метода unique()
print('Проверим полученную категорию, с помощью метода unique():')
print(data['Образование'].unique())
# Вычислим количество соискателей со средним уровенем образования (школьное образование)
print('Количество соискателей со средним уровенем образования (школьное образование):', 
      data[data['Образование'] == 'Среднее']['Образование'].count())
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# Объявляем функцию которая принимает строку и возвращает первое слово перед первой запятой (пол)
def gender(feature):
    """Функция-преобразования для нового признака 'Пол'

    Args:
        feature (str): значение должно браться из колонки 'Пол, возраст'

    Returns:
        str: первое слово перед запятой от входного значения (пол)
    """
    # Полученный входной агрумент освобождаем от точек и запятых, приводим к нижнему регистру и
    # делим на слова через пробелы, заносим в список слов 
    lst = feature.replace(',','').replace('.', '').lower().split(' ')
    # В цикле идем по списку входящих слов
    for index, word in enumerate(lst):
        # Проверяем в списке входных слов наличие: жен* и муж*
        if len(word) >= 3:
            if word[0:3] in ['жен', 'муж']:
                # Если найдено 'жен' или 'муж' тогда выводим первый символ из текущего элемента списка 
                return lst[index][0].upper()

    return lst[0].rstrip()
# Объявляем функцию которая принимает строку и возвращает числовое значенние после первой запятой (возраст)
def age(feature):
    """Функция-преобразования для нового признака 'Возраст'

    Args:
        feature (str): значение должно браться из колонки 'Пол, возраст'

    Returns:
        int: числовое значение после первой запятой от входного значения (возраст)
    """
    # Полученный входной агрумент освобождаем от точек и запятых, приводим к нижнему регистру и
    # делим на слова через пробелы, заносим в список слов 
    lst = feature.replace(',','').replace('.', '').lower().split(' ')
    # В цикле идем по списку входящих слов
    for index, word in enumerate(lst):
        # Проверяем в списке слов входных наличие слов: год* и лет
        if len(word) >= 3:
            if word[0:3] in ['год', 'лет']:
                # Если найдено 'год' или 'лет' тогда выводим возраст из предыдущего элемента списка 
                return lst[index - 1]

# Создаем с помощью функции-преобразования новый признак 'Пол'
data['Пол'] = data['Пол, возраст'].apply(gender)
# Переобразовываем признак 'Пол' в категориальный тип данных
data['Пол'] = data['Пол'].astype('category')
# Создаем с помощью функции-преобразования новый признак 'Возраст'
data['Возраст'] = data['Пол, возраст'].apply(age)
# Переобразовываем признак 'Образование' в числовой тип данных
data['Возраст'] = data['Возраст'].astype('int')
# Удаляем колонку 'Пол, возраст'
data = data.drop('Пол, возраст', axis=1)
# Вычислим процент женских резюме представленых в наших данных
print('Процент женских резюме представленых в наших данных: {}%'.format(
         round(data[data['Пол']=='Ж']['Пол'].count() / data.shape[0] * 100, 2)))
# Вычислим средний возраст соискателей
print('Cредний возраст соискателей:', round(data['Возраст'].mean(),1))
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Объявляем функцию, которая вычесляет количество месяцев из строкового выражения,
# содержащего выражения, например: Опыт работы 2 года 1 месяц…; Опыт работы 6 месяцев…
def experience(feature):
    """Функция-преобразования для нового признака 'Опыт работы (месяц)'

    Args:
        feature (str): значение должно браться из колонки 'Опыт работы'

    Returns:
        int: вычисляемое количество месяцев 
    """
    # Если значание пустое или 'Не указано', тогда возвращаем np.nan
    if feature is np.nan or feature == 'Не указано':
        return np.nan
    # Делим входной агрумент на элементы разделенные пробелом, получаем список 
    lst = feature.split()
    month = 0
    for index, word in enumerate(lst):
        # Ишем ключевые слова 'год' или 'лет'
        if len(word) >= 3:
            if word[0:3] == 'год' or word[0:3] == 'лет':
                # вычисляем кол-во месяцев путем умножения предыдущего элемента списка на 12
                month = int(lst[index-1]) * 12
                # если после кол-ва лет не указано кол-во месяцев, 
                # то выводим результат
                if lst[index+2][0:3] != 'мес':
                    return month
            # ищем ключевое слово 'мес'
            if word[0:3] == 'мес':
                # если найдено, 
                # тогда прибавляем его к текущему количеству месяцев и выводим результат
                return month + int(lst[index-1])
# Создаем с помощью функции-преобразования новый признак 'Опыт работы (месяц)'
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(experience)
# Удаляем колонку 'Опыт работы'
data = data.drop('Опыт работы', axis=1)
# Вычеслим медианный опыт работы (в месяцах) в нашей таблице
print('Медианный опыт работы (в месяцах):', data['Опыт работы (месяц)'].median())
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Объявляем функцию которая определяет категорию города из признака 'Город, переезд, командировки'
def City(feature):
    """Функция-преобразования для нового признака 'Город'

    Args:
        feature (str): значение должно браться из колонки 'Город, переезд, командировки'

    Returns:
        str: вычисляемая категория города 
    """
    # Объявляем список городов милионников
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород',
                    'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
                    'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    # Разбиваем входной агрумент на список, берем первый элемент из списка, обрезаем пробел справа
    City = feature.split(',')[0].rstrip()
    # Возвращаем название города без изменений если это Москва и СПБ
    if City == 'Москва' or City == 'Санкт-Петербург':
        return City
    # Возвращаем город-милионник если город относится к списку городов милионников
    elif City in million_cities:
        return 'город-миллионник'
    # Возвращаем катекогию другие если не сработали условия выше
    else:
        return 'другие'

# Объявляем функцию которая определяет готовность к переезду из признака 'Город, переезд, командировки'
def Willing_to_relocate(feature):
    # Разбиваем входной агрумент на список, 
    # берем срез списка от второго до предпоследнего элемента включительно, 
    # в данном диапазоне находится признак 'Готовность к переезду'
    lst_features = feature.split(',')[1:-1]
    # Создаем цикл по словам в цикле по признакам 
    for feature in lst_features:
        lst_words = feature.split()
        for word in lst_words:
            # По ключевым словам находим нужный признак
            if len(word) >=6:
                if word[0:6] == 'переез' or word[0:6] == 'переех':
                    # Возвращаем обратный результат поиска слова 'не' в нужном нам признаке  
                    return not 'не' in lst_words

# Объявляем функцию которая определяет готовность к переезду из признака 'Город, переезд, командировки'
def willingness_to_travel(feature):
    # Разбиваем входной агрумент на список, 
    # берем срез списка - последний элемент,
    # в котором должен находится признак 'Готовность к командировкам'
    lst_features = feature.split(',')
    lst_words = lst_features[-1].split()
    # Создаем цикл по словам в признаке 'Готовность к командировкам'
    for word in lst_words:
        # Определяем соответствие признака по ключевому слову
        if len(word) >=10:
            if word[0:10] == 'командиров':
                # Возвращаем обратный результат поиска слова 'не' в нужном нам признаке  
                return not 'не' in lst_words
    # Если информация о готовности к командировкам не указана, то соискатель не готов к командировкам.
    return False

# Создаем с помощью функций-преобразования новые признаки:
# 'Город', 'Готовность к переезду', 'Готовность к командировкам'
data['Город'] = data['Город, переезд, командировки'].apply(City)
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(Willing_to_relocate)
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(willingness_to_travel)
# Переобразовываем новые признакм в категориальный тип данных
data['Город'] = data['Город'].astype('category')
# Удаляем колонку 'Город, переезд, командировки'
data = data.drop('Город, переезд, командировки', axis=1)
# Вычислим процент соискателей, которые живут в Санкт-Петербурге
print('Процент соискателей, которые живут в Санкт-Петербурге:',
      round(data[data['Город'] == 'Санкт-Петербург']['Город'].count() / data.shape[0] * 100))
# Вичислим процент соискателей, которые готовы одновременно и к переездам, и к командировкам
print('Процент соискателей, которые готовы одновременно и к переездам, и к командировкам:', 
      round(data[(data['Готовность к переезду']==True) & (data['Готовность к командировкам']==True)].count()[0] /
             data.shape[0] * 100))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Объявляем функцию которая определяет наличие подпризнака в признаке
# Например наличие подпризнака 'полная занятость' в признаке 'Занятость'
# Или наличие подпризнака 'полный день' в признаке 'График'
def employment_and_schedule(feature, subfeature):
    """Функция-преобразования для множества новых признаков,
      названия которых состоят в признаках 'Занятость' и 'График'

    Args:
        feature (str): значение должно браться из колонок 'Занятость' и 'График'
        subfeature (str): в данном аргументе передается название подпризнака

    Returns:
        boolean: возращает наличие подпризнака (subfeature) в признаке (feature)
    """
    # Получаем список подпризнаков
    lst = feature.split(',')
    # Убираем пробелы слева и справа, приводим к нижнему регистру все элементы списка
    lst = [s.strip().lower() for s in lst]
    # Возвращаем результат наличия подпризнака в списке подпризнаков 
    return subfeature in lst

# Объявляем вспомогательную функцию для формирования множества вариантов
# подпризнаков в интерисующем нас признаке,
# например в признаке 'График', могут быть найдены сл. подпризнаки:
# 'Вахтовый метод', 'Гибкий график', 'Полный день', 'Сменный график', 'Удаленная работа'
def get_subfeatures_in_feature(feature):
    """вспомогательная функция возвращающая множество свовосочетаний
    встречающихся в признаке указанном во входном параметре, 
    в датасете data

    Args:
        feature (str): наименование признака

    Returns:
        list: отсортированный лист из множества вариантов словосочетаний 
    """
    # объявляем пустое множество
    subfeatures = set()
    # идем по списку уникальных значений всречающихся в заданном признаке 
    for lst_features in data[feature].unique():
        # делим уникальный вариант заданного признака на словосочетания (подпризнаки) 
        for feature in lst_features.split(','):
            # добавляем во множество найденный подпризнак, 
            # предварительно обрезав пробелы слева и справа и преобразовав нижний регистр 
            subfeatures.add(feature.strip().lower())
    # возвращаем отсортированное множество ввиде списка
    return sorted(subfeatures)

# Запускаем цикл из двух признаков 'Занятость' и 'График'
for feature in ['Занятость', 'График']:
    # Подучаем список подпризнаков из текущего признака 
    subfeatures = get_subfeatures_in_feature(feature)
    # Циклически создаем новые подпризнаки с помощью вспомогательной функции 
    for subfeature in subfeatures:
        data[subfeature[0].upper()+subfeature[1:]] = data[feature].apply(employment_and_schedule,
                                                                    subfeature=subfeature)

# Удаляем колонки 'Занятость' и 'График'
data = data.drop(['Занятость', 'График'], axis=1)

# Вычислим количество людей, которые ищут проектную работу и волонтёрство
print('Количество людей, которые ищут проектную работу и волонтёрство:',
       data[data['Проектная работа'] & data['Волонтерство']].count().values[0])

# Вычислим количество людей, которые хотят работать вахтовым методом и с гибким графиком 
print('Количество людей, которые хотят работать вахтовым методом и с гибким графиком:', 
      data[data['Вахтовый метод'] & data['Гибкий график']].count().values[0])
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Объявляем функцию которая определяет временный признак 'Валюта' из признака 'ЗП'
def currency(feature):
    # Объявляем словарь сопоставляющий наименования валют в основном датасете с
    # наименованиям валют по стандарту ISO (из выгрузки валют)
    currency_names = {
                'грн': 'UAH', 'USD': 'USD', 'EUR': 'EUR', 'белруб': 'BYN',
                'KGS': 'KGS', 'сум': 'UZS', 'AZN': 'AZN', 'KZT': 'KZT'
                }
    # возвращаем наименование курса валют по стандарту ISO,
    # в качестве ключа берем второй элемент (наименование валюты) из разбитого входного аргумента 
    return currency_names.get(feature.split()[1].replace('.',''))

# Объявляем функцию которая создает новый признак 'ЗП (руб)' из признаков 'ЗП', 'close' и 'proportion''
def salary_RUB (features):
    # Проверяем если, признак 'close' пустой, 
    # т.е. данные из выгрузки валют не подгрузились, значит ЗП в рублях,
    # то возращаем просто признак 'ЗП'
    if np.isnan(features)[1]:
        return features[0]
    # Иначе это ЗП, которая указана в иностранной валюте,
    # возращаем вычисляемое значение: 'ЗП' * 'close' / 'proportion' 
    else:
        return features[0] * features[1] / features[2]
    
# Считываем новый датасет - выгрузка валют
ExchangeRates = pd.read_csv('data/ExchangeRates.csv')
# Преобразовываем признак 'date' во временной тип данных 
ExchangeRates['date'] = pd.to_datetime(ExchangeRates['date'])
# Удаляем неиспользуемые колонки 'per', 'time', 'vol' из выгрузки валют
ExchangeRates = ExchangeRates.drop(['per', 'time', 'vol'], axis=1)

# Создаем с помощью функции-преобразования временный признак 'Валюта'
data['Валюта'] = data['ЗП'].apply(currency)
# Преобразовываем с помощью функции-преобразования признак 'ЗП'
data['ЗП'] = data['ЗП'].apply(lambda x: float(x.split()[0]))
# Преобразовываем признак 'ЗП' во числовой тип данных 
#data['ЗП'] = data['ЗП'].astype('int')
# Преобразовываем признак 'Обновление резюме' во временной тип данных 
data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме']).dt.date.astype('datetime64[ns]')
# Объединяем основной датасет с выгразками валют по двум полям 
data = data.merge(ExchangeRates, how='left', left_on=['Валюта', 'Обновление резюме'], 
          right_on=['currency', 'date'])
# Создаем с помощью функции-преобразования новый признак 'ЗП (руб)'
data['ЗП (руб)'] = data[['ЗП', 'close', 'proportion']].apply(salary_RUB, axis=1)

# Удаляем неиспользуемые колонки 'ЗП', 'Валюта', 'vol' из основного датасета
data = data.drop(['ЗП', 'Валюта', 'currency', 'date', 'close', 'proportion'], axis=1)

# Вычисляем желаемую медианную заработную плату соискателей в нашей таблице (в рублях)
print('Желаемая медианная заработная плата соискателей: {} тыс. рублей'.
      format(round(data['ЗП (руб)'].median() / 1000)))
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Объявляем функцию определения выбросов по методу Z-отклонений
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    """фукция предназначена для определения выбросов по методу Z-отклюнений

    Args:
        data (DataFrame): исходный датасет
        feature (str): рассматриваемый признак
        log_scale (bool, optional): Если указано True, то признак берется в логорифмическом масштабе. По умолчанию False.
        left (int, optional): Число сигм, которое берется в левую сторону. По умодчанию 3.
        right (int, optional): Число сигм, которое берется в правую сторону. По умодчанию 3.

    Returns:
        outliers, cleaned (DataFrame): возращается два результирующих датасета (только с выбросами и без выбросов)
    """

    # Вычисляем признак в логарифмическом масштабе, если это необходимо
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    
    # Вычисляем среднее значение
    mu = x.mean()
    # Вычисляем стандартное отклонение
    sigma = x.std()
    # Вычисляем нижний и верхний пределы 
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    # Получаем датасеты только с выбросами и без выбросов
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    # Возвращаем результирующие датасеты
    return outliers, cleaned

# Формируем канвас для гистограммы
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
# Поучаем признак в логорифмирующем масштабе
log_age = np.log(data['Возраст'] + 1)
# Формируем гистограмму
histplot = sns.histplot(log_age, bins=30, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 4 * log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение возраста в логарифмическом маштабе')
histplot.set_ylabel('Количество соискателей');
```

```python
# Вызываем функцию по определению z-отклонений, получаем датасеты с только выбросами и без выбросов 
outliers, cleaned = outliers_z_score_mod(data, 'Возраст', log_scale=True, left=3, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
# Выводим таблицу с полученными выбросами
display(outliers)
# Удаляем в основном датасете выбросы полученные по методу z-отклонений
data = cleaned
```
